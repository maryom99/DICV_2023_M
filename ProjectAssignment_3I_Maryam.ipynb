## AVG Filter

```python
def convolution2d(image, kernel, stride, padding):
  
    # Pad the input image with zeros
    image_padded = np.pad(image, ((padding, padding), (padding, padding)), 'constant')
  
    # Define the output image size based on the input image size, kernel size, and padding
    output_shape = ((image.shape[0] - kernel.shape[0] + 2*padding)//stride + 1, 
                    (image.shape[1] - kernel.shape[1] + 2*padding)//stride + 1)
  
    # Initialize the output image
    output = np.zeros(output_shape)
  
    # Iterate over the input image with the given stride
    for i in range(0, image_padded.shape[0] - kernel.shape[0] + 1, stride):
        for j in range(0, image_padded.shape[1] - kernel.shape[1] + 1, stride):
      
            # Define the region of interest (ROI)
            roi = image_padded[i:i+kernel.shape[0], j:j+kernel.shape[1]]
      
            # Apply the kernel to the ROI
            k = (roi * kernel).sum()
      
            # Store the convolved pixel in the output image
            output[i//stride, j//stride] = k
  
    return output


# Load the input image
img = cv.imread('/content/drive/MyDrive/images_DICV/R.jpeg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Define the kernel
kernel_avg = (1/9) * np.array([[1, 1, 1],[1, 1, 1],[1, 1, 1]])
# Apply the convolution
output = convolution2d(img_gray, kernel_avg, 1, 2)

# Display the output image
cv2_imshow(output)
```
